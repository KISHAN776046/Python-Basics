```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate dataset
np.random.seed(0)
categories = ['A', 'B', 'C']
values = np.concatenate([np.random.normal(loc, 1, 100) for loc in [5, 10, 15]])
cat = np.repeat(categories, 100)

df = pd.DataFrame({'Category': cat, 'Value': values})

# Boxplot to visualize distribution across categories
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Distribution of Values Across Categories')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate correlated data
np.random.seed(0)
x1 = np.random.randn(100)
x2 = 0.8 * x1 + np.random.randn(100) * 0.2
x3 = np.random.randn(100)
df = pd.DataFrame({'Feature1': x1, 'Feature2': x2, 'Feature3': x3})

# Compute correlation matrix
corr = df.corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Heatmap for Sales Data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot table for heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".0f")
plt.title('Sales Variation by Month and Day')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('3D Surface Plot: z = sin(sqrt(x² + y²))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10,6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*5, alpha=0.6, color='coral', edgecolors='black')
plt.title('Bubble Chart: GDP vs Population')
plt.xlabel('GDP')
plt.ylabel('Population')

for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], country, fontsize=9, ha='center', va='center')

plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

df = pd.DataFrame({
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
})

mapper = LinearColorMapper(palette="Viridis256", low=df.value.min(), high=df.value.max())

p = figure(title="Heatmap", x_range=(0,1), y_range=(0,1), tools="hover", tooltips=[("Value", "@value")], match_aspect=True)
p.rect(x='x', y='y', width=0.1, height=0.1, source=df,
       fill_color=transform('value', mapper), line_color=None)

color_bar = ColorBar(color_mapper=mapper, location=(0,0))
p.add_layout(color_bar, 'right')

show(p)
```
